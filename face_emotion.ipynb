```python
import cv2
import matplotlib.pyplot as plt
```

```python
img =  cv2.imread("happy1.jpg") #loading image

#showing image with plot 
plt.imshow(img)   
# cv2.waitKey(0)
```

```python
#intially all the images will be in BGR so if we want in color(RGB) here is the code
color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow( color_img)
```

```python
#importing deepface library and DeepFace
from deepface import DeepFace


#this analyses the given image and gives values
#1st time using this, it may give many errors and some google drive links to download some .h5 and zip files, download and save them in the location where it shows that files are missing.
prediction = DeepFace.analyze(img)
```

```python
#lets check what all the things DeepFace.analyze() function has analyzed 
prediction
```

### dominant_emotion and dominant_race are our predicted values by DeepFace.analyze() function by analyzing the picture/image

```python
#lets see what emotion does the model predicted

prediction['dominant_emotion']
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
```

```python
# gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(color_img, 1.1, 4)

for (x, y, u, v) in faces:
    cv2.rectangle(color_img, (x,y), (x+u, y+v), (0, 0, 225), 2)
```

```python
plt.imshow(color_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(color_img, prediction['dominant_emotion'], (0, 50), font, 1, 
(225,0,0), 2, cv2.LINE_4)
cv2.putText(color_img, prediction['dominant_race'], (40, 100),font,  0.5,
(0,0,0), 2, cv2.LINE_4)

plt.imshow(color_img)

# plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
